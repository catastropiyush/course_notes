```python
def round_to_sig_figs(number, sig_figs):
    if number == 0:
        return 0
    return round(number, sig_figs - int(math.floor(math.log10(abs(number)))) - 1)


#function to calculate different types of errors
def calculate_errors(true_value, approximate_value, significant_figures):

    rounded_approx = round_to_sig_figs(approximate_value, significant_figures)

    true_error = true_value - rounded_approx

    absolute_error = abs(true_error)

    if true_value != 0:
        relative_true_error = true_error / true_value
    else:
        relative_true_error = float('inf')

    percentage_error = relative_true_error * 100



    true_error          = round_to_sig_figs(true_error, significant_figures)
    absolute_error      = round_to_sig_figs(absolute_error, significant_figures)
    relative_true_error = round_to_sig_figs(relative_true_error, significant_figures)
    percentage_error    = round_to_sig_figs(percentage_error, significant_figures)

    return true_error, absolute_error, relative_true_error, percentage_error
```

```python
true_value          = 453.16852
approximate_value   = 453.1685
significant_figures = 7

true_error, absolute_error, relative_true_error, percentage_error = calculate_errors(
    true_value, approximate_value, significant_figures)

print(f"True Error             : {true_error}")
print(f"Absolute Error         : {absolute_error}")
print(f"Relative True Error    : {relative_true_error}")
print(f"Percentage Error       : {percentage_error} %")
```

```python
def f(x):
    return x**5 - 8*x**4 + 44*x**3 - 91*x**2 + 85*x - 26

def bisection(f, a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("Function must have opposite signs at a and b")

    steps = []
    for _ in range(max_iter):
        c = (a + b) / 2
        steps.append((a, b, c))
        if f(c) == 0 or (b - a) / 2 < tol:
            return c, steps
          #If the function values at c and a have opposite signs, then the root must lie between c and b or else betwen a and c
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c
    raise ValueError("Method did not converge")

a, b = 0, 1                      # intervals
root, steps = bisection(f, a, b) #
print(f"Root = {root:.6f}")

x = np.linspace(-1, 3,100)
y = f(x)

plt.figure(figsize=(5,5))
plt.plot(x, y, label=r'$f(x) = x^5 - 8x^4 + 44x^3 - 91x^2 + 85x - 26$', color='deeppink', linewidth=2)
plt.axhline(y =  0, color='grey', linestyle='--')
plt.axvline(x =root, color='black', linestyle='--', label='Root')
plt.plot(root, 0, 'r*', markersize=15)

plt.xlabel('x')
plt.ylabel('f(x)')

plt.title('Bisection Method')
plt.legend()
plt.grid(True, alpha=0.3)

plt.xlim(0, 2.5)
plt.ylim(-3, 3)

plt.tight_layout()
plt.show()

# Print steps
for i, (left, right, mid) in enumerate(steps):
    print(f"Step {i+1}: a = {left:.4f}, b = {right:.4f}, c = {mid:.4f}, f(a)= {f(left):.4f}, f(b)={f(right):.4f}, f(c) = {f(mid):.4f}")
```

```python
def false_position(func, a, b, tolerance=1e-6, max_iterations=100):
  if func(a) * func(b) >= 0:
    print("Function values at the interval endpoints must have opposite signs.")
    return None, []

  iterations = []
  for i in range(max_iterations):
    c = (a * func(b) - b * func(a)) / (func(b) - func(a))
    iterations.append((i + 1, a, b, c, func(c)))

    if abs(func(c)) < tolerance:
      return c, iterations

    if func(c) * func(a) < 0:
      b = c
    else:
      a = c

  print("Maximum number of iterations reached.")
  return c, iterations

def f(x):
  return 4*x**3 - 6*x**2 + 7*x -2.3

a = 0
b = 1.5
root, iterations = false_position(f, a, b)

if root is not None:
  print("Approximate root:", root)
  x_vals = np.linspace(a - 1, b + 1, 400)
  y_vals = f(x_vals)
  plt.plot(x_vals, y_vals, label="f(x)")
  plt.plot(root, 0, 'r*', markersize=15)
  plt.xlabel("x")
  plt.ylabel("f(x)")
  plt.title("False Position Method")
  plt.legend()
  plt.grid(True)
  plt.show()

  print("Iteration | a | b | f(c)")
  for iteration in iterations:
    print("{:<9} | {:.6f}|{:.6f}|{:.6f}".format(*iteration))
```

$= \frac{\mathrm{d}}{\mathrm{d}x}\left[4 - x\right] \cdot \mathrm{e}^{-\frac{x}{2}} + \left(4 - x\right) \cdot \frac{\mathrm{d}}{\mathrm{d}x}\left[\mathrm{e}^{-\frac{x}{2}}\right] + 0$
$= \left(\frac{\mathrm{d}}{\mathrm{d}x}\left[4\right] - \frac{\mathrm{d}}{\mathrm{d}x}\left[x\right]\right) \mathrm{e}^{-\frac{x}{2}} + \left(4 - x\right) \mathrm{e}^{-\frac{x}{2}} \cdot \frac{\mathrm{d}}{\mathrm{d}x}\left[-\frac{x}{2}\right]$
$= \left(0 - 1\right) \mathrm{e}^{-\frac{x}{2}} + \left(4 - x\right) \mathrm{e}^{-\frac{x}{2}} \left(-\frac{1}{2} \cdot \frac{\mathrm{d}}{\mathrm{d}x}\left[x\right]\right)$
$= -\frac{\left(4 - x\right) \mathrm{e}^{-\frac{x}{2}} \cdot 1}{2} - \mathrm{e}^{-\frac{x}{2}}$
$= -\frac{\left(4 - x\right) \mathrm{e}^{-\frac{x}{2}}}{2} - \mathrm{e}^{-\frac{x}{2}}$

```python
def f(x):
    return (4-x)*np.exp(-0.5*x)-2

def f_prime(x):
    return 0.5*((x-6)*np.exp(-0.5*x))

def newtoroot(f, df, x0, tol=1e-5, max_iter=100):
    for _ in range(max_iter):
        if abs(f(x0)) < tol:
            return x0
        x0 = x0 - f(x0) / df(x0)
        print(f"x={x0:.4f},df(x)={df(x0):.4f},f(x)={f(x0):.4f}")
    raise ValueError("Method did not converge")

x0 = -1.3
root = newtoroot(f, f_prime, x0)
print(f"Root = {root:.6f}")

x = np.linspace(-5, 5, 200)
y = f(x)

plt.figure(figsize=(5,5))
plt.plot(x, y, label=r'$f(x)$', color='deeppink', linewidth=2)
plt.plot(x, f_prime(x), label=r'$\frac{df}{dx}$', color='blue', linewidth=2)
plt.axhline(y=0, color='grey', linestyle='--')
plt.axvline(x=root, color='black', linestyle='--')
plt.plot(root, 0, 'r*', markersize=15, label='Root')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Newton-Raphson Method')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(-5, 5)
plt.ylim(-5, 10)
```

```python
def f(x):
    return 2*np.sin(np.sqrt(x))-x

def g(x):
    return 2*np.sin(np.sqrt(x))

def fixed_point_iteration(x0, tol=1e-6, max_iter=50):
    x = x0
    iterations = [x]
    for i in range(max_iter):
        x_new = g(x)
        iterations.append(x_new)
        if abs(x_new - x) < tol:
            return iterations
        x = x_new
    return iterations

starting_points = [0.5]
fig, ax = plt.subplots(figsize=(5,7))

x = np.linspace(-3,5,500)
ax.plot(x, f(x), 'b-', label='$f(x)$')
ax.plot(x, g(x), 'g-', label='$g(x)$')
ax.plot(x, x, 'k--', label='y = x', alpha=0.5)

colors = ['r', 'm', 'c']
for x0, color in zip(starting_points, colors):
    iterations = fixed_point_iteration(x0)

    print(f"\nIterations for starting point x₀ = {x0}:")
    for i, x in enumerate(iterations):
        print(f"  x_{i} = {x:.8f}")

    # Plot iterations
    ax.plot(iterations, iterations, f'{color}o-', alpha=0.6,
            label=f'Start: {x0}')

    for i in range(len(iterations) - 1):
        ax.plot([iterations[i], iterations[i], iterations[i+1]],
                [iterations[i], g(iterations[i]), g(iterations[i])],
                f'{color}--', linewidth=0.8, alpha=0.6)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("Fixed-Point Iteration")
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)
ax.set_xlim(-2, 4)
ax.set_ylim(-2, 6)

plt.tight_layout()
plt.show()
print(f"\nFinal approximation of the root: {iterations[-1]:.8f}")
```

```python
def f(x):
    return 2 * np.sin(np.sqrt(x)) - x

def g(x):
    return 2 * np.sin(np.sqrt(x))

def wegstein(g, x0, x1, tol=1e-6, max_iter=100):
    path = [x0, x1]
    for i in range(max_iter):
        x2 = g(x1)
        if abs(x2 - x1) < tol:
            path.append(x2)
            return x2, path
        s = (x2 - g(x0)) / (x1 - x0)
        if s == 1:           #slope is parallel,no convergence
            x_next = x2
        else:
            x_next = (x2 - s*x1) / (1 - s)
        path.append(x_next)
        x0, x1 = x1, x_next
    raise ValueError("Method did not converge")

x0, x1 = 0.5, 1.29

root, path = wegstein(g, x0, x1)
print(f"Root = {root:.6f}")

x = np.linspace(0, 3, 300)
y = f(x)

plt.figure(figsize=(5,5))
plt.plot(x, y, label='f(x) = 2sin(√x) - x', color='deeppink', linewidth=2)
plt.plot(x, g(x), label='g(x) = 2sin(√x)', color='blue', linewidth=1)
plt.axhline(y=0, color='grey', linestyle='--')
plt.axvline(x=root, color='black', linestyle='--')
plt.plot(root, 0, 'r*', markersize=15, label='Root')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Wegstein Method')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 3)
plt.ylim(-1, 2)

plt.show()

# Print iteration path
print("\nIteration path:")
for i, xi in enumerate(path):
    print(f"Step {i}: x = {xi:.6f}, f(x) = {f(xi):.6f}")
```

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

def f(x):
    return np.array([
        x[0]**2 + x[1] - 11,  # x^2 + y = 11
        x[1]**2 + x[0] - 7    # y^2 + x = 7
    ])

def jacobian(x):
    return np.array([
        [2*x[0], 1],
        [1, 2*x[1]]
    ])

def newton_method(f, jacobian, x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        if np.linalg.norm(fx) < tol:
            return x
        J = jacobian(x)
        delta = np.linalg.solve(J, -fx)
        x = x + delta
    raise RuntimeError("did not converge")

def visualize_system(f, root, x_range=(-5, 5), y_range=(-5, 5)):
    x = np.linspace(*x_range, 200)
    y = np.linspace(*y_range, 200)
    X, Y = np.meshgrid(x, y)
    Z1 = f([X, Y])[0]
    Z2 = f([X, Y])[1]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4))
    im1 = ax1.imshow(Z1, cmap='viridis', extent=(*x_range, *y_range), origin='lower', interpolation='bilinear')
    ax1.contour(X, Y, Z1, levels=[0], colors='r', linewidths=2)
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_title(r'$x^{2} + y - 11 = 0$')
    fig.colorbar(im1, ax=ax1)

    im2 = ax2.imshow(Z2, cmap='plasma', extent=(*x_range, *y_range), origin='lower', interpolation='bilinear')
    ax2.contour(X, Y, Z2, levels=[0], colors='r', linewidths=2)
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_title(r'$y^{2} + x - 7 = 0$')
    fig.colorbar(im2, ax=ax2)

    ax1.plot(root[0], root[1], 'r*', markersize=10)
    ax2.plot(root[0], root[1], 'r*', markersize=10)
    plt.tight_layout()
    plt.show()

initial_guess = np.array([1.0, 2.0])
root = newton_method(f, jacobian, initial_guess)

print("Root found by Newton's method:", root)
print("Function value at root:", f(root))


root_scipy = fsolve(f, initial_guess)
print("Root found by scipy's fsolve:", root_scipy)
print("Function value at scipy's root:", f(root_scipy))

visualize_system(f, root)
```

```python
def gaussian_elimination(A, b):
    n = len(b)
    Ab = np.column_stack((A, b))#augmented matrix

    def visualize_matrix(matrix, step):
        plt.figure(figsize=(5,5))
        im = plt.imshow(np.flipud(matrix), cmap='coolwarm') # Flip the matrix vertically to correct orientation
        #plt.colorbar(im)
        plt.title(f"Matrix at step {step}")

        for i in range(matrix.shape[0]):
            for j in range(matrix.shape[1]):
                text = plt.text(j, matrix.shape[0] - 1 - i, f'{matrix[i, j]:.2f}',ha='center', va='center', color='w')
        plt.yticks(range(matrix.shape[0]), range(matrix.shape[0], 0, -1))
        plt.tight_layout()

    visualize_matrix(Ab, "Initial")

    # Forward elimination
    for i in range(n):
        # Find pivot row
        max_element = abs(Ab[i:, i]).argmax() + i
        if Ab[max_element, i] == 0: #If the pivot element is zero,
                                    #it means the matrix is singular and cannot be inverted.
            raise ValueError("Matrix is singular.")

        # Swap maximum element row with current row
        Ab[i], Ab[max_element] = Ab[max_element], Ab[i].copy()

        # Make all rows below this one 0 in current column
        for j in range(i + 1, n):
            factor = Ab[j, i] / Ab[i, i]
            Ab[j, i:] -= factor * Ab[i, i:]

        visualize_matrix(Ab, f"Elimination step {i+1}")

    # Back substitution
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (Ab[i, -1] - np.dot(Ab[i, i+1:n], x[i+1:])) / Ab[i, i]

    visualize_matrix(Ab, "Final")

    return x
```

```python
def gauss_jordan_elimination(A, b):
    n = len(b)
    Ab = np.column_stack((A, b))

    def visualize_matrix(matrix, step):
        plt.figure(figsize=(5,5))
        im = plt.imshow(np.flipud(matrix), cmap='coolwarm')
        #plt.colorbar(im)
        plt.title(f"Matrix at step {step}")

        for i in range(matrix.shape[0]):
            for j in range(matrix.shape[1]):
                text = plt.text(j, matrix.shape[0] - 1 - i, f'{matrix[i, j]:.2f}',ha='center', va='center', color='w')
        plt.yticks(range(matrix.shape[0]), range(matrix.shape[0], 0, -1))
        plt.tight_layout()
    visualize_matrix(Ab, "Initial")

    for i in range(n):
        max_element = abs(Ab[i:, i]).argmax() + i
        if Ab[max_element, i] == 0:
            raise ValueError("Matrix is singular.")

        #swap current row with max row
        Ab[i], Ab[max_element] = Ab[max_element], Ab[i].copy()
        # Normalize the pivot row
        pivot = Ab[i, i]
        Ab[i, :] /= pivot

        for j in range(n):
            if j != i:
                factor = Ab[j, i]
                Ab[j, :] -= factor * Ab[i, :]
        visualize_matrix(Ab, f"Elimination step {i+1}")

    #solution is in the last column
    x = Ab[:, -1]
    visualize_matrix(Ab, "Final")
    return x
```

```python
A = np.array([
    [2, 1, -1],
    [5, 2, 2],
    [3, 1, 1]
], dtype=float)
b = np.array([1,-4,5], dtype=float)

try:
    solution = gauss_jordan_elimination(A, b)
    print("Solution:", solution)
except ValueError as e:
    print(str(e))
```

```python
def gauss_seidel(A, b, x0, max_iterations=100, tolerance=1e-6):
  n = len(b)
  x = x0.copy()

  for iteration in range(max_iterations):
    x_new = x.copy()
    for i in range(n):
      sigma = 0
      for j in range(n):
        if i != j:
          sigma += A[i, j] * x_new[j]
      x_new[i] = (b[i] - sigma) / A[i, i]
      print(x_new)
    if np.linalg.norm(x_new - x) < tolerance:
      return x_new
    x = x_new

  print("Gauss-Seidel did not converge")
  return None

A = np.array([
    [10, 2, -1],
    [-3,-6,  2],
    [1,  1,  5]
], dtype=float)
b = np.array([27,-61.5,-21.5], dtype=float)

# Initial guess
x0 = np.array([0.0,0.0,0.0])

solution = gauss_seidel(A, b, x0,max_iterations=50,tolerance=1e-6)
if solution is not None:
  print("Solution (Gauss-Seidel):", solution)
```
